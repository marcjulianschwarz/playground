# The need for hidden layers (XOR function)

```python
import numpy as np
from mbnn import plot_one, plot_three, plot_two

def two_d_point(xmin, xmax, ymin, ymax, step):
    x = np.arange(xmin, xmax, step)
    y = np.arange(ymin, ymax, step)
    return [(i, j) for i in x for j in y]


def f1(p):
    x, y = p
    return np.array([-10, 10]) @ (np.array([x, y]) - np.array([0, 0.5]))


def f2(p):
    x, y = p
    return np.array([10, -10]) @ (np.array([x, y]) - np.array([0.5, 0]))

points = two_d_point(0, 1, 0, 1, 0.1)
y1 = [f1(i) for i in points]
y2 = [f2(i) for i in points]

xs, ys = zip(*points)

# Adding points (0,0), (1,0), (0,1) and (1,1) for both functions
important_points = [(0, 0), (1, 0), (0, 1), (1, 1)]
important_y1 = [f1(i) for i in important_points]
important_y2 = [f2(i) for i in important_points]

important_xs, important_ys = zip(*important_points)

plot_one(xs, ys, y1, y2, important_xs, important_ys, important_points, f1, f2)
plot_two(xs, ys, y1, y2, f1, f2)
plot_three(f1, f2)
```
