# Proof Sketch of the Universal Approximation Theorem

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def step(x):
    return 1 if x >= 0 and x <= 1 else 0
```

```python
x = np.arange(-2.0, 2.1, 0.1)
y = [i**2 for i in x]
ai = [np.random.randn() for _ in range(7)]
bi = [np.random.randn() for _ in range(7)]
ci = [np.random.randn() for _ in range(7)]

# Weierstrass
def sin(x):
    return sum([ai[i] * np.tanh(bi[i] * x + ci[i]) for i in range(7)])

# Fourier
def step_sin(x):
    return sum([ai[i] * sin(bi[i] * x + ci[i]) for i in range(7)])

# Weierstrass
def f(x):
    return sum([ai[i] * step_sin(bi[i] * x + ci[i]) for i in range(7)])


y2 = [f(i) for i in x]
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
