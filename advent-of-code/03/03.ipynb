```python
from pathlib import Path 

test_data = Path("test.txt").read_text()
data = Path("data.txt").read_text()
```

```python
data = [[val for val in line] for line in data.splitlines()]
```

```python
def is_symbol(s: str):
    return not s.isdigit() and not s == "."
```

```python
n = len(data) - 1
m = len(data[0]) - 1
n, m
```

```python
total_part_sum = 0

for i, row in enumerate(data):
    current_digit = ""
    has_symbol = False

    for j, el in enumerate(row):
        if el.isdigit():
            current_digit += el
            # print(i, j, current_digit, has_symbol)

            # right
            if j < m and is_symbol(row[j + 1]):
                has_symbol = True
            # left
            if j > 0 and is_symbol(row[j - 1]):
                has_symbol = True
            # down
            if i < n and is_symbol(data[i + 1][j]):
                has_symbol = True
            # up
            if i > 0 and is_symbol(data[i - 1][j]):
                has_symbol = True
            # diagonal right down
            if i < n and j < m and is_symbol(data[i + 1][j + 1]):
                has_symbol = True
            # diagonal left up
            if i > 0 and j > 0 and is_symbol(data[i - 1][j - 1]):
                has_symbol = True
            # diagonal left down
            if i < n and j > 0 and is_symbol(data[i + 1][j - 1]):
                has_symbol = True
            # diagonal right up
            if i > 0 and j < m and is_symbol(data[i - 1][j + 1]):
                has_symbol = True

        if not el.isdigit() or j == m:
            if current_digit:
                if has_symbol:
                    total_part_sum += int(current_digit)
            current_digit = ""
            has_symbol = False
print(total_part_sum)
```

```python
gear_boxes = []
for i, row in enumerate(data):
    for j, el in enumerate(row):
        if el == "*":
            gear_box = []

            # top
            if i > 0:
                gear_box.append(
                    data[i - 1][j - 3 : j + 4] if j >= 3 else data[i - 1][: j + 4]
                )
            # right and left
            gear_box.append(row[max(0, j - 3) : j + 4])
            # bottom
            if i < n:
                gear_box.append(
                    data[i + 1][j - 3 : j + 4] if j >= 3 else data[i + 1][: j + 4]
                )
            gear_boxes.append(gear_box)
```

```python
def get_gear_value(gear_box):
    def is_symbol(s: str):
        return s == "*"

    connected_gears = []
    n = len(gear_box) - 1
    m = len(gear_box[0]) - 1

    for i, row in enumerate(gear_box):
        current_digit = ""
        is_gear_connected = False

        for j, el in enumerate(row):
            if el.isdigit():
                current_digit += el
                # right
                if j < m and is_symbol(row[j + 1]) and j+1==3:
                    is_gear_connected = True
                # left
                if j > 0 and is_symbol(row[j - 1]) and j-1==3:
                    is_gear_connected = True
                # down
                if i < n and is_symbol(gear_box[i + 1][j]) and i+1==1 and j==3:
                    is_gear_connected = True
                # up
                if i > 0 and is_symbol(gear_box[i - 1][j]) and i-1==1 and j==3:
                    is_gear_connected = True
                # diagonal right down
                if i < n and j < m and is_symbol(gear_box[i + 1][j + 1]) and i+1==1 and j+1==3:
                    is_gear_connected = True
                # diagonal left up
                if i > 0 and j > 0 and is_symbol(gear_box[i - 1][j - 1]) and i-1==1 and j-1==3:
                    is_gear_connected = True
                # diagonal left down
                if i < n and j > 0 and is_symbol(gear_box[i + 1][j - 1]) and i+1==1 and j-1==3:
                    is_gear_connected = True
                # diagonal right up
                if i > 0 and j < m and is_symbol(gear_box[i - 1][j + 1]) and i-1==1 and j+1==3:
                    is_gear_connected = True

            if not el.isdigit() or j == m:
                if current_digit:
                    if is_gear_connected:
                        connected_gears.append(int(current_digit))
                current_digit = ""
                is_gear_connected = False
    #print(connected_gears)
    if len(connected_gears) == 2:
        return connected_gears[0] * connected_gears[1]
    else:
        return 0
```

```python
gear_box_sum = 0
for gear_box in gear_boxes:
    gear_box_sum += get_gear_value(gear_box)
```
